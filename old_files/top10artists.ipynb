## <center>Dataset</center>

```python
# Importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Connecting the JSON file to Pandas
df = pd.read_json('AccountData/StreamingHistory0.json')
df.to_csv('SpotifyStreamData.csv')
df
```

```python
# Cleaning the time column to the correct format.
df['playTime'] = pd.to_datetime(df['endTime'])
df['year'] = pd.DatetimeIndex(df['playTime']).year
df['month'] = pd.DatetimeIndex(df['playTime']).month
df['day'] = pd.DatetimeIndex(df['playTime']).day
df['weekday'] = pd.DatetimeIndex(df['playTime']).weekday
df['time'] = pd.DatetimeIndex(df['playTime']).time
df['hours'] = pd.DatetimeIndex(df['playTime']).hour
df['day-name'] = df['playTime'].apply(lambda x: x.day_name())
df['artists_count'] = 1
df['songs_count'] = 1
df['timePlayed'] = pd.to_timedelta(df['msPlayed'], unit='ms')
df['ListenTimeHRS'] = df['timePlayed'].apply(lambda x: x.seconds/3600).round(3)
df['ListenTimeMNTS'] = df['timePlayed'].apply(lambda x: (x.seconds/60)%60).round(3)
df.drop(columns=['endTime', 'timePlayed', 'msPlayed'], inplace=True)
```

```python
df
```

## <center>Artists</center>

```python
# Removing music before 1st Jan 2023
df['playTime'] = pd.to_datetime(df['playTime'])
lower_bound = pd.to_datetime('2022-12-31')
upper_bound = pd.to_datetime('2023-12-01')
artists_df = df[(df['playTime'] > lower_bound) & (df['playTime'] < upper_bound)].reset_index().drop(columns=['index'])
artists_df
```

```python
# Checking unique values
artists_df.nunique()
```

```python
# Calculating the unique artists
total_artists = artists_df['artistName'].count().sum()
unique_artist = artists_df['artistName'].nunique()
total_artists, unique_artist
```

```python
# Percentage of unique artists
plt.rcParams["font.family"] = "Arial"
uniqueArtistList = np.array([unique_artist, total_artists - unique_artist])
unique_artistLab = ['Unique Artists', 'Non-Unique Artists']
colors = ['#ADFF2F', '#FFA500']
explode = (0.05, 0.05)
plt.pie(uniqueArtistList, colors=colors, labels=unique_artistLab, autopct='%1.1f%%', pctdistance=0.85, explode=explode, startangle=180)
centre_circle = plt.Circle((0, 0), 0.7, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Percentage of Unique Artists')
plt.savefig('unique_artists.png')
plt.show()
```

```python
# Grouping the artists and sorting them for top 10 artists
top10ArtistsTimeDF = artists_df.groupby(['artistName'])[['ListenTimeHRS', 'ListenTimeMNTS', 'artists_count']].sum().sort_values(by=['ListenTimeHRS','artists_count'], ascending=False).reset_index().head(10)
top10ArtistsTimeDF
```

```python
# Calculate the percentage of total per artist
total_listen_time = top10ArtistsTimeDF['ListenTimeHRS'].sum()
top10ArtistsTimeDF['percentage_total_artists'] = (top10ArtistsTimeDF['ListenTimeHRS'] / total_listen_time) * 100

# Set the style for Seaborn
sns.set(style="whitegrid")

# Create a bar chart
plt.figure(figsize=(10, 6))
bar_chart = sns.barplot(x='artistName', y='ListenTimeHRS', hue='percentage_total_artists', data=top10ArtistsTimeDF, palette='viridis', dodge=False)
bar_chart.set_xticks(range(len(top10ArtistsTimeDF)))
bar_chart.set_xticklabels(top10ArtistsTimeDF['artistName'], rotation=45, horizontalalignment='right')
bar_chart.set_title('Top 10 Artists Wrapped from Spotify')
bar_chart.set_xlabel('Artist Name')
bar_chart.set_ylabel('Listen Time (Hours)')

# Remove the legend as we are using percentage directly on the bars
bar_chart.legend_.remove()

# Add percentage labels over the bars
for index, value in enumerate(top10ArtistsTimeDF['percentage_total_artists']):
    bar_chart.text(index, top10ArtistsTimeDF['ListenTimeHRS'].iloc[index] + 0.1,
                   f'{value:.1f}%', ha='center', va='center', fontsize=8, color='black')

# Remove the gridlines
plt.grid(False)

# Save the graph to a PNG file
plt.savefig('top10_artists.png')

# Display the chart
plt.show()
```

## <center>Songs</center>

```python
top10SongsTimeDF = artists_df.groupby(['trackName'])[['ListenTimeHRS', 'ListenTimeMNTS', 'songs_count']].sum().sort_values(by=['songs_count','ListenTimeHRS'], ascending=False).reset_index().head(10)
top10SongsTimeDF
```

```python
# Calculating the %age for Songs Total
total_songs_count = top10SongsTimeDF['songs_count'].sum()
top10SongsTimeDF['percentage_count_songs'] = (top10SongsTimeDF['songs_count'] / total_songs_count) * 100

# Set the style for Seaborn
sns.set(style="whitegrid")

# Create a bar chart
plt.figure(figsize=(10, 6))
bar_chart = sns.barplot(x='trackName', y='songs_count', data=top10SongsTimeDF, palette='viridis', hue='percentage_count_songs', dodge=False, legend=False)
bar_chart.set_xticks(range(len(top10SongsTimeDF)))
bar_chart.set_xticklabels(top10SongsTimeDF['trackName'], rotation=45, horizontalalignment='right')
bar_chart.set_title('Top 10 Songs Wrapped from Spotify')
bar_chart.set_xlabel('Track Name')
bar_chart.set_ylabel('No. of Times Songs Played')

# Add percentage labels over the bars
for index, value in enumerate(top10SongsTimeDF['percentage_count_songs']):
    bar_chart.text(index, top10SongsTimeDF['songs_count'].iloc[index] + 0.1,
                   f'{value:.1f}%', ha='center', va='bottom', fontsize=8, color='black')

# Remove the gridlines
plt.grid(False)

# Save the graph to a PNG file
plt.savefig('top10_songs.png')

# Display the chart
plt.show()
```
